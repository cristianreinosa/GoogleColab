```python
import pandas as pd
```

# Nueva sección

```python
datos_viviendas = pd.read_csv("alquiler.csv", sep = ";")
```

```python
datos_viviendas
```

```python
sel_departamentos = datos_viviendas["Tipo"] == "Departamento"
cant_dpto = datos_viviendas[sel_departamentos].shape[0]
cant_dpto
```

```python
sel_otros_departamentos = datos_viviendas["Tipo"] == "Departamento"
cant_otra_vivi = datos_viviendas[~sel_otros_departamentos].shape[0]
cant_otra_vivi
```

```python
sel_casa_hab_cochera = (datos_viviendas["Tipo"] == "Casa") | (datos_viviendas["Tipo"] == "Habitación") | (datos_viviendas["Tipo"] == "Cochera")
cant_dpto_casa_hab_cochera = datos_viviendas[sel_casa_hab_cochera].shape[0]
cant_dpto_casa_hab_cochera
```

```python
sel_area = (datos_viviendas["Area"] >= 60) & (datos_viviendas["Area"] <= 100)
cant_viviendas = datos_viviendas[sel_area].shape[0]
cant_viviendas
```

```python
sel_hab_precio = (datos_viviendas["Cuartos"] >= 4) & (datos_viviendas["Valor"] <= 2000 )
cant_hab_precio = datos_viviendas[sel_hab_precio].shape[0]
cant_hab_precio
```

```python
print("Se encontraron los siguientes resultados:")
print(f"El numero de dptos es de {cant_dpto}")
print(f"El numero de viviendas restantes es de:  {cant_otra_vivi}")
print(f"El numero de viviendas tipo casa,habitacion y cochera: {cant_dpto_casa_hab_cochera}")
print(f"La cantidad de viviendas con un area entre 60 y 100 metros cuadrados es de  {cant_viviendas}")
print(f"La cantidad de viviendas con almenos 4 hab y con un valor menir o igual a 2000 es : {cant_hab_precio}")
```
