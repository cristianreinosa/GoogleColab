#Nueva seccion

```python
datos_pc = pd.read_excel('laptops.xlsx')
```

```python
datos_pc
```

```python
sel_role = datos_pc['Role'] == 'Office Manager'
cant_back = datos_pc[sel_role].shape[0]
cant_back
```

```python
sel_marca_pc = datos_pc['Purchased For'] == 'Medellin'
cant_otro_purs = datos_pc[sel_marca_pc].shape[0]
cant_otro_purs
```

```python
sel_marca_pc = (datos_pc['Brand'] == 'Lenovo') | (datos_pc['Role']== 'Frond end Developer')
cat_otro_marca = datos_pc[sel_marca_pc].shape[0]
cat_otro_marca
```

```python
sel_role_model = (datos_pc['Role'] == 'BI Engineer') | (datos_pc['Model'] == 'Lenovo E14 i5')
cant_role_model = datos_pc[sel_role_model].shape[0]
cant_role_model
```

```python
sel_Ai_engineer = datos_pc['Role'] == 'AI Engineer'
cant_ai_engineer = datos_pc[sel_Ai_engineer].shape[0]
cant_ai_engineer
```

```python
print("Se encontraron los siguientes resultados")
print(f"En el rol manager se encontraron un total de:{cant_back}")
print(f"El numero de contratacion en ciudad de medellin se encontro un total de:{cant_otro_purs}")
print(f"El numero de marca Lenovo y rol frond end deloper se encontro un total de:{cat_otro_marca} ")
print(f"El totoal de rol BI Enginner y Modelo lenovo E14 i5 es de un total:{cant_role_model}")
print(f"El numero de rol de AI Enginner es de :{cant_ai_engineer}")
```

```python
roles = ['Office Manager', 'Medellin', 'Lenovo/Front end Developer', 'BI Engineer/Modelo Lenovo E14 i5', 'AI Engineer']
cantidades = [cant_back, cant_otro_purs, cat_otro_marca, cant_role_model, cant_ai_engineer]

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(roles, cantidades, color=['blue', 'green', 'orange', 'red', 'purple'])
plt.xlabel('Roles/Marcas')
plt.ylabel('Cantidad de Laptops')
plt.title('Cantidad de Laptops por Rol/Marca')
plt.show()
```
